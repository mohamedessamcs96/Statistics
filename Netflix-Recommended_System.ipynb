### Finding Similaritites 

```python
import pandas as pd
```

```python
ratings=pd.read_csv('Netflix_Dataset_Rating.csv')
ratings.head()
```

```python
movies=pd.read_csv('Netflix_Dataset_Movie.csv')
movies.head()
```

```python
movie_ratings=pd.merge(movies,ratings)
```

```python
movie_ratings.head()
```

```python
movie_ratings.shape
```

```python
movie_ratings=movie_ratings.pivot_table(index=['User_ID'],columns=['Name'],values='Rating')
movie_ratings.head()
```

```python
movies['Name'][movies['Name'].str.contains('Beautiful ')]
```

```python
beautifulMind=movie_ratings['A Beautiful Mind']
beautifulMind.head()
```

```python
similar_movies=movie_ratings.corrwith(beautifulMind)
similar_movies=similar_movies.dropna()

df=pd.DataFrame(similar_movies)

df.head(10)
```

```python
similar_movies.sort_values(ascending=False)
```
